```
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black');
```

```
Y_F=np.array(([0.53,0.93,0.88,0.78,0.77,0.96,0.71,0.66,0.69,0.99]))
Y_CR=np.array(([0.33,0.45,0.15,0.09,0.50,0.37,0.18,0.09,0.17,0.30]))
Num=len(Y_F)
X_F=np.zeros(Num)
for i in range(Num):
  X_F[i]=(i+1)/Num
print(Y_F)
print(X_F)
```

```
plt.plot(X_F, Y_F, 'o', color='black');
```

```
plt.plot(X_F, Y_F,'go--', linewidth=2, markersize=12)
```

```
plt.plot(X_F, Y_F, 'g^', X_F, Y_CR, 'g-')
```

```
plt.plot(X_F, Y_F, 'go--', X_F, Y_CR, 'g^-')
```

```
plt.plot(X_F, Y_F, 'go--', X_F, Y_CR, 'g^-', linewidth=2, markersize=12)
```

```
plt.plot(X_F, Y_F, 'ko--', X_F, Y_CR, 'k^-', linewidth=2, markersize=12)
```

```
from matplotlib import pyplot as PLT
fig = PLT.figure()
ax1 = fig.add_subplot(111)
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)
plt.setp(ax1.spines.values(), linewidth=2.0)
plt.plot(X_F, Y_F, 'ko--', X_F, Y_CR, 'k^-', linewidth=2, markersize=12)
ax1.grid(False)
```

```
from pylab import *
from matplotlib import pyplot as PLT
fig = PLT.figure()
ax1 = fig.add_subplot(111)
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)
plt.setp(ax1.spines.values(), linewidth=2.0)
plt.plot(X_F, Y_F, 'ko--', X_F, Y_CR, 'k^-', linewidth=2, markersize=12)
ax1.grid(False)
# Change size and font of tick labels
# Again, this doesn't work in interactive mode.
fontsize = 14
ax = gca()

for tick in ax.xaxis.get_major_ticks():
    tick.label1.set_fontsize(fontsize)
    tick.label1.set_fontweight('bold')
for tick in ax.yaxis.get_major_ticks():
    tick.label1.set_fontsize(fontsize)
    tick.label1.set_fontweight('bold')

xlabel('Fraction intearations', fontsize=16, fontweight='bold')
ylabel('F and CR', fontsize=16, fontweight='bold')
#ax.legend( ('F', 'CR'), loc='upper right')
ax.legend( ('F', 'CR'), loc=[0.8,0.4])
```
