```
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
from pylab import *
from matplotlib import pyplot as PLT
```

```
F_Rastrigin=np.array(([0.5,0.53,0.93,0.88,0.78,0.77,0.96,0.71,0.66,0.69,0.99]))
CR_Rastrigin=np.array(([0.5,0.33,0.45,0.15,0.09,0.50,0.37,0.18,0.09,0.17,0.30]))
Num=len(F_Rastrigin)
X_F=np.zeros(Num)
for i in range(Num):
  X_F[i]=(i)/Num
```

```
f = plt.figure()

Y_CR=CR_Rastrigin
Y_F=F_Rastrigin
fig = PLT.figure()
ax1 = fig.add_subplot(111)
ax1.set_xlim(0, 1)
ax1.set_ylim(0, 1)
plt.setp(ax1.spines.values(), linewidth=2.0)
plt.plot(X_F, Y_F, 'ko--', X_F, Y_CR, 'k^-', linewidth=2, markersize=12)
ax1.grid(False)
# Change size and font of tick labels
# Again, this doesn't work in interactive mode.
fontsize = 14
ax = gca()

for tick in ax.xaxis.get_major_ticks():
    tick.label1.set_fontsize(fontsize)
    tick.label1.set_fontweight('bold')
for tick in ax.yaxis.get_major_ticks():
    tick.label1.set_fontsize(fontsize)
    tick.label1.set_fontweight('bold')

xlabel('Fraction intearations', fontsize=16, fontweight='bold')
ylabel('F and CR', fontsize=16, fontweight='bold')
#ax.legend( ('F', 'CR'), loc='upper right')
ax.legend( ('F', 'CR'), loc=[0.8,0.4])

f.savefig( "Rastrigin_jun_04_2020.png")
```
